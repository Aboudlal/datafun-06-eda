# Notebook starter (first cells)

```python
import pandas as pd

df = pd.read_csv("tips.csv")  # file is at repo root
df["tip_pct"] = df["tip"] / df["total_bill"]
df.head()
```

# Basic checks:

```python
df.info()
df.describe(include="all")
df.isna().sum()

```

# Quick charts (matplotlib):

```python
import matplotlib.pyplot as plt

df["total_bill"].plot(kind="hist", bins=20, title="Total Bill Distribution"); plt.show()
df["tip_pct"].plot(kind="hist", bins=20, title="Tip % Distribution"); plt.show()

(df.groupby("day")["tip_pct"].mean().sort_values()
  .plot(kind="bar", title="Average Tip % by Day")); plt.show()

(df.groupby(["smoker","time"])["tip_pct"].mean()
  .unstack().plot(kind="bar", title="Average Tip % by Smoker & Time")); plt.show()
```
